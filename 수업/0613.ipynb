```python
X = np.array([1,2,3,4,5,6])
y = np.array([5,8,11,14,17,20])
# y = 3*x + 2
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(units = 10, input_dim = 1))
model.add(keras.layers.Dense(units = 1))
```

## 문제

```python
# 입력 1~100까지 
# 출력 : 2배 + 1
optimizer = 'adam'
loss = 'mse'
metrics = 'mse'
epoch = 10000
# 200, 300, 400의 값을 예측해 봅시다.
```

```python
# 데이터 만들기
# y = 2*x + 1
X = np.arange(1, 101)
y = 2*X + 1

#모델 만들기 (model->레이어 추가)
model = keras.Sequential()
model.add(keras.layers.Input(shape = (1,)))
model.add(keras.layers.Dense(units=10)) #10 = unit(동그라미) 의미함/input값이 몇차원인지 알려줘야함
model.add(keras.layers.Dense(units=1))

#컴파일
model.compile(optimizer = 'adam',
              loss = 'mae',
              metrics = ['mae'])

#모델 훈련
model.fit(X, y, epochs = 100)

# 훈련 평가
model.evaluate(X, y)

# 예측
model.predict(np.array([200, 300, 400]))

#array([[401.01013],
#       [601.01666],
#       [801.0232 ]], 
```
